```python
a = read_cdf(files[0])
_, bins = pd.qcut(a['mass_values'], 1000, duplicates='drop',retbins=True)
```

```python
def binning(a, bins=None, sparse=None):
    """
    结构化色质联用数据
    """
    
    _cdf = pd.DataFrame({'mz': a['mass_values'], 'int': a['intensity_values']})
    if bins is None:
        _, bins = pd.cut(_cdf.mz, 1000, duplicates='drop',retbins=True)
    scan_grp = pd.cut(a.point_number+1,
                       np.append(a['scan_index'], a.point_number.max()+1),
                       include_lowest=True)
    mz_grp = pd.cut(_cdf.mz, bins)
    scans = _cdf.groupby([scan_grp, mz_grp])
    matrix = scans.int.sum().unstack('mz')
    if sparse:
        matrix = matrix.fillna(0).to_sparse(fill_value=0)
    return matrix
```

```python
df = binning(a, bins, sparse=False)
```
