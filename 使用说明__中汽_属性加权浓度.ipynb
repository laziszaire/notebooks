## 数据准备
1. 在桌面上新建一个文件夹，命名为tests
2. 将各个样本的浓度测试结果如{顶棚.xlsx,地毯.xlsx}放入tests文件夹内
3. 将组分属性信息表命名为properties.xlsx，之后放到桌面上
4. 将‘车内空气VOCs评价’文件夹中的‘**中汽_属性加权浓度.ipynb**'放到桌面
5. 找到桌面的位置如，C:\Users\litao\Desktop


## 安装软件
1. 车内空气VOCs评价文件夹中找到 Anaconda3-4.4.0-Windows-x86_64.exe
    双击，然后点几次下一步

## 运行软件
1. 按windows键+R, 然后数据cmd打开命令窗口
2. 输入jupyter notebook 你的桌面位置 
如：
~~~
    jupyter notebook C:\Users\litao\Desktop
    ~~~
3. 点击打开‘中汽_属性加权浓度.ipynb’
4. 在上方工具栏点击kernel, 然后点击Restart & Run All


## 结果
桌面上的 **data.xls**，即为程序运行结果
