```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
logC = np.linspace(1,10,20)
C = np.exp(logC)
```

```python
a, b = -2, 3
y = a*logC + b
```

```python
plt.semilogx(C,y, '-o')
plt.xlabel('C')
plt.title('')
```

```python
plt.plot(logC, y, '-o')
plt.xlabel('logC')
```

```python
plt.plot(C, y, '-o')
plt.xlabel('C')
```
