```python
np.random.seed(1)
f = plt.figure(figsize=(15,8))
colors = np.random.random(len(chromatograms)-1)
for chromatograph,color in zip(chromatograms[1:],colors):
    plt.scatter(chromatograph['time array'],np.tile(chromatograph['Q3'],chromatograph['time array'].shape),
                c=np.tile(color,chromatograph['time array'].shape),alpha=.1,cmap=plt.cm.Accent,vmin=0,vmax=1)
plt.xlabel('retension time')
plt.ylabel('Q3')
plt.savefig('3d_xic.png',dpi=300)
```
